```python
# Make sure the plot shows up
%matplotlib inline

import aiohttp
import asyncio
import aiomoex
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns

sns.set()

# Graphics in SVG format are more sharp and legible
%config InlineBackend.figure_format = 'svg'


async def main():
    async with aiohttp.ClientSession() as session:
        data1 = await aiomoex.get_board_candles(session, 'GMKN')
        data2 = await aiomoex.get_board_candles(session, 'SBER')
        df1 = pd.DataFrame(data1)
        df2 = pd.DataFrame(data2)
        #df.set_index('begin', inplace=True)
        #print(df.head(), '\n')
        #print(df.tail(), '\n')
        return df1, df2
df1, df2 = await main()
```

```python
def get_returns(data):
    """
    This function get_data reads a data file from disk
    and returns percentage returns.
    """
    return pd.DataFrame(data).pct_change()
```

```python
#функция корреляции
def calc_corr(ser1, ser2, window):
    """
    Calculates correlation between two series.
    """
    ret1 = ser1.pct_change()
    ret2 = ser2.pct_change()
    corr = ret1.rolling(window).corr(ret2)
    return corr
```

```python
df1.head(10)
```

```python
# Define how many points we intend to plot. Points in this case would be trading days.
points_to_plot = 300
df3 = df1.merge(df2, on='begin', how='left')
df_final = df3[['begin', 'close_x', 'close_y']]
df_final.set_index('begin', inplace=True)
df_final = df_final.rename(columns={'close_x': 'GMKN', 'close_y': 'SBER'})

#Преобразуем "2024-03-01 00:00:00" в "2024-03-01"
df_final.index = pd.to_datetime(df_final.index).date
```

```python
df_final
```

```python
# Rebase the two series to the same point in time, starting where the plot will start.
for i in df_final:
    df_final[i + '_rebased'] = (df_final[-points_to_plot:][i].pct_change() + 1).cumprod()
```

```python
df_final.tail(10)
```

```python
sns.jointplot(x=df_final.index, y="GMKN", data=df_final, kind="scatter")
```

```python
sns.distplot(df_final["SBER"])
```

```python
# Relative strength, GMKN to SBER
df_final['rel_str'] = df_final['GMKN'] / df_final['SBER']
# Calculate 50 day rolling correlation
df_final['corr'] = calc_corr(df_final['GMKN'], df_final['SBER'], 100)
```

```python
# Slice the data, cut points we don't intend to plot.
plot_data = df_final[-points_to_plot:]

# Make new figure and set the size.
fig = plt.figure(figsize=(12, 8))

# The first subplot, planning for 3 plots high, 1 plot wide, this being the first.
ax = fig.add_subplot(311)
ax.set_title('GMKN-SBER Comparison')
ax.semilogy(plot_data['GMKN_rebased'], linestyle='-', label='GMKN', linewidth=3.0)
ax.semilogy(plot_data['SBER_rebased'], linestyle='--', label='SBER', linewidth=3.0)
ax.legend()
ax.grid(False)

# Second sub plot.
ax = fig.add_subplot(312)
ax.plot(plot_data['rel_str'], label='Relative Strength, GMKN to SBER', linestyle=':', linewidth=3.0)
ax.legend()
ax.grid(True)

# Third subplot.
ax = fig.add_subplot(313)
ax.plot(plot_data['corr'], label='Correlation between GMKN and SBER', linestyle='-.',
linewidth=3.0)
ax.legend()
ax.grid(True)
```
