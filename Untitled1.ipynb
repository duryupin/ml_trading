def get_returns(df2):
    """
    This function get_data reads a data file from disk
    and returns percentage returns.
    """
    return pd.DataFrame(df2).pct_change()

#функция корреляции
def calc_corr(ser1, ser2, window):
    """
    Calculates correlation between two series.
    """
    ret1 = ser1.pct_change()
    ret2 = ser2.pct_change()
    corr = ret1.rolling(window).corr(ret2)
    return corr

df2.tail(10)

# Define how many points we intend to plot. Points in this case would be trading days.
points_to_plot = 300
df3 = df1.merge(df2, on='begin', how='left')
df_final = df3[['begin', 'close_x', 'close_y']]
df_final.set_index('begin', inplace=True)
df_final = df_final.rename(columns={'close_x': 'GMKN', 'close_y': 'SBER'})

#Преобразуем "2024-03-01 00:00:00" в "2024-03-01"
df_final.index = pd.to_datetime(df_final.index).date

df_final

# Rebase the two series to the same point in time, starting where the plot will start.
for i in df_final:
    df_final[i + '_rebased'] = (df_final[-points_to_plot:][i].pct_change() + 1).cumprod()

df_final.tail(10)

sns.jointplot(x=df_final.index, y="SBER", data=df_final, kind="scatter")

sns.displot(df_final["SBER"], )
plt.show()

# Relative strength, GMKN to SBER
df_final['rel_str'] = df_final['GMKN'] / df_final['SBER']
# Calculate 100 day rolling correlation
df_final['corr'] = calc_corr(df_final['GMKN'], df_final['SBER'], 100)

df_final.tail(10)

plot_data = df_final[-points_to_plot:]
